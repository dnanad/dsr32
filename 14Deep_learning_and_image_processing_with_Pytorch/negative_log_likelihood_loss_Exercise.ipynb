```python
import numpy as np

#index 0 corresponds to cat
#index 1 corresponds to dog
#index 2 corresponds to horse
cat_prediction_logits = np.array([5, 4, 2])
horse_prediction_logits = np.array([4, 2, 8])
dog_cat_prediction_logits = np.array([4, 4, 1])
dog_prediction_logits = np.array([10, 1, 3])
```

```python
import math
def softmax(vec):
  exps = np.array([math.exp(val) for val in vec])
  return exps / sum(exps)
```

```python
softmax(cat_prediction_logits)
```

```python
softmax(horse_prediction_logits)
```

```python
# this needs to add up to one
np.sum(softmax(cat_prediction_logits))
```

```python
def nll(softmax_vec):
  index_of_max = np.argmax(softmax_vec)
  return -math.log(softmax_vec[index_of_max])
```

```python
nll(softmax(cat_prediction_logits))
```

```python
nll(softmax(horse_prediction_logits))
```

```python
nll(softmax(dog_cat_prediction_logits))
```
