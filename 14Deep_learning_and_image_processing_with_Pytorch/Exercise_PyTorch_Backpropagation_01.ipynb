```python
import torch
```

```python
?torch.tensor
```

```python
# suppose this is a feature
x = torch.tensor(1.0)

# and this is a true value 
y = torch.tensor(2.0)
x, y
```

```python
# the prediction here is simply going to be a weight, times the feature
# Notice that requires_grad is False by default
# check ?torch.tensor
# to backprop should we have requires_grad be True or False? 
# What are we tracking with the 'gradient tape'? 
w = torch.tensor(1.0, requires_grad=True)
w
```

```python
# forward pass
y_hat = w * x
y_hat
```

```python
#Q: Notice the changes in grad_fn, why do you think this is? 
#Q2: How would grad_fn change if we delete **2?
#Q3: How would grad_fn change if we use torch.abs(y_hat -y)? Which loss function would this be?    
loss = (y_hat - y) ** 2
loss
```

```python
# compute gradient, should you do it on loss or on y_hat? 
loss.backward()
```

```python
# only w should have an accessible gradient 
w.grad
```
